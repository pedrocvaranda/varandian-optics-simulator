# Varandian Optics - Quick Start

Interactive demonstration of **Varandian Optics**: light propagation in curved spaces.

Based on the paper:
> *Varandian Optics: A Non-Euclidean Formulation of Light Propagation*  
> Pedro Coutinho Varanda, Zenodo, 2026

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('..')

from core import compute_ray, SK, space_type_name
from visualization import plot_single_ray, plot_comparison

# Matplotlib settings
plt.rcParams['figure.dpi'] = 100
%matplotlib inline
```

## Example 1: Single Ray in Spherical Space (K > 0)

Positive curvature - light follows great circles!

```python
# Parameters
r0 = 0.5      # Initial radial position
theta0 = 0    # Initial angle
K = 1.0       # Curvature (positive = spherical)
C = 0.3       # Constant of motion

# Compute ray
theta, r = compute_ray(r0=r0, theta0=theta0, K=K, C=C, method='numerical')

# Plot
print(f"Computing ray in {space_type_name(K)} space...")
fig, ax = plot_single_ray(theta, r, K)
plt.show()

print(f"\nRay statistics:")
print(f"  Points computed: {len(theta)}")
print(f"  r_min = {r.min():.4f}")
print(f"  r_max = {r.max():.4f}")
```

## Example 2: Hyperbolic Space (K < 0)

Negative curvature - visualized in Poincaré disk!

```python
# Hyperbolic parameters
K_hyp = -1.0
C_hyp = 0.4

# Compute
theta_hyp, r_hyp = compute_ray(r0=0.5, theta0=0, K=K_hyp, C=C_hyp)

# Plot
print(f"Computing ray in {space_type_name(K_hyp)} space...")
fig, ax = plot_single_ray(theta_hyp, r_hyp, K_hyp)
plt.show()
```

## Example 3: Comparison Across Curvatures

Same initial conditions, different spaces!

```python
# Common parameters
r0_common = 0.6
theta0_common = 0
C_common = 0.35

# Compute for different K
K_values = [-1.0, 0.0, 1.0]
results = {}

for K_val in K_values:
    theta_temp, r_temp = compute_ray(r0=r0_common, theta0=theta0_common, 
                                      K=K_val, C=C_common)
    results[K_val] = r_temp

# Same theta for all
theta_common = theta_temp

# Plot comparison
fig, axes = plot_comparison(theta_common, results, K_values=K_values, figsize=(16, 5))
plt.show()

print("Notice how the same ray behaves VERY differently in each space!")
```

## Example 4: Verifying the Metric (Equation 2)

Let's verify that $S_K(r)$ behaves as expected from the paper.

```python
# Test SK function
r_test = np.linspace(0.01, 2.0, 100)

SK_sphere = SK(r_test, K=1.0)
SK_eucl = SK(r_test, K=0.0)
SK_hyp = SK(r_test, K=-1.0)

plt.figure(figsize=(10, 6))
plt.plot(r_test, SK_sphere, 'b-', linewidth=2, label='Spherical (K=+1)')
plt.plot(r_test, SK_eucl, 'g-', linewidth=2, label='Euclidean (K=0)')
plt.plot(r_test, SK_hyp, 'r-', linewidth=2, label='Hyperbolic (K=-1)')

plt.xlabel('r', fontsize=12)
plt.ylabel('$S_K(r)$', fontsize=12)
plt.title('Metric Function $S_K(r)$ - Equation (2) from Paper', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Spherical: SK(r) = (1/√K) sin(√K*r) - bounded!")
print("Euclidean: SK(r) = r - linear")
print("Hyperbolic: SK(r) = (1/√|K|) sinh(√|K|*r) - exponential growth!")
```
