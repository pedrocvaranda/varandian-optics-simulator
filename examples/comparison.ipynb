```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('..')

from core import compute_ray, space_type_name
from visualization import plot_comparison

plt.rcParams['figure.dpi'] = 100
%matplotlib inline
```

## Comparison 1: Same Initial Conditions, Different Curvatures

Let's see how the SAME ray behaves in spherical, Euclidean, and hyperbolic spaces.

```python
# Common parameters
r0 = 0.6
theta0 = 0
C = 0.35

# Compute for all three geometries
K_values = [-1.0, 0.0, 1.0]
results = {}

for K in K_values:
    print(f"Computing ray in {space_type_name(K)} space (K={K})...")
    theta, r = compute_ray(r0=r0, theta0=theta0, K=K, C=C)
    results[K] = r

theta_common = theta  # Same for all

print("\nComputations complete!")
```

```python
# Plot comparison
fig, axes = plot_comparison(theta_common, results, K_values=K_values, figsize=(18, 5))
plt.show()

print("\nKey Observation:")
print("The SAME initial conditions produce drastically different trajectories!")
print("This demonstrates how geometry fundamentally shapes light propagation.")
```

## Comparison 2: Varying the Constant C

Now let's fix the geometry and vary the conserved quantity C.

```python
plt.show()

# Define a fixed geometry for this comparison (change as desired)
K_fixed = 0.0
print(f"\nIn {space_type_name(K_fixed)} space:")
print("Larger C -> tighter geodesics")
print("This reflects the 'angular momentum' conservation in curved space.")
```

## Comparison 3: Metric Behavior

Let's visualize how SK(r) behaves in different spaces.

```python
plt.show()

print("\nMetric Behavior:")
print("• Spherical: Bounded, oscillatory (sin behavior)")
print("• Euclidean: Linear growth")
print("• Hyperbolic: Exponential growth (sinh behavior)")
print("\nThis fundamentally determines how 'space expands' as you move radially.")
```

## Key Takeaways
